```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sps
```

```python
x = np.arange(0,1,0.01)
mu = lambda x: 0.5-(x)**2
sig = lambda x: 0.02+0.2*x**2
# sig = lambda x: 0.0001
```

```python
fig, ax = plt.subplots()

plt.plot(x,mu(x),'b')
plt.plot(x,mu(x)+sig(x),'b--')
plt.plot(x,mu(x)-sig(x),'b--')

# ax.set_aspect('equal')
ax.grid(True, which='both')

ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

plt.xlim(0,1)
```

```python
fig, ax = plt.subplots()

plt.plot(x,1/2*(1-sps.erf(-mu(x)/(sig(x)/10))),'b')

ax.grid(True, which='both')

ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

plt.xlim(0,1)
plt.ylim(0,1.05)
```
