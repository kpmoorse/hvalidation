```python
import numpy as np
import sklearn.gaussian_process as skg
import sklearn as sk
import matplotlib.pyplot as plt
```

```python
K = skg.kernels.ConstantKernel(1., constant_value_bounds="fixed") * skg.kernels.RBF(2, length_scale_bounds="fixed")
gp0 = skg.GaussianProcessRegressor()
x = np.arange(-1,1,0.01).reshape(-1,1)
```

```python
for i in range(10):
    ys = gp0.sample_y(x,random_state=None)
    plt.plot(x,ys)
```
