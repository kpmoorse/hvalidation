```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gauss(x,mu=0,sigma=1):
    a = 1/np.sqrt(2*np.pi*sigma**2)
    b = (x-mu)**2/(2*sigma**2)
    return a*np.exp(-b)
```

```python
h = 0.01
x = np.arange(-5,5,h)
plt.plot(x,gauss(x))
plt.plot(x,x**2*gauss(x))
plt.show()
```

```python
x = np.arange(0.01,200,h)
np.sum(np.sqrt(x)*gauss(x,mu=100,sigma=0.1))*h
```
