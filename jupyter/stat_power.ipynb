```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sps
```

```python
def split_binom(p,n,x,a=0.8):
    
    return sps.comb(n,x)*(a*p**x*(1-p)**(n-x) + (1-a)*.5**n)
```

```python
n=100
eps=0.1
p0=0.95
pdf = split_binom(p0,n,np.arange(n+1),a=0.5)
plt.plot(pdf)
plt.show()
print(1-np.sum(pdf[:int(np.floor(n-eps*n)+1)]))
```
